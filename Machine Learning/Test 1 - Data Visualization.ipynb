```python
# X and y arrays (It should be 2D array)

X = df[['Population']]           # X = np.array (df['Population']) Convert from 1D to 2D
y = df['Deaths - Cumulative']    # X_2d = np.reshape(X, (30, 1))                 

# We separated our data into X and y where X is the features and y is what we're trying to predict.
# X = Training set 
# y= Test set 
```

```python
#  Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
# Creating and Training the Model

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
df.dtypes
```

```python
#Model Evaluation

# print the intercept
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df

# Holding all other features fixed, a 1 unit increase in Population is associated with an *increase of 0.000788 *.
```

```python
# Predictions from our Model

predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
# y= aX + b 
# y = death 
# a = Coefficient = 0.000788
# X = population
# b = intercept = -3.87

# Y = (0.000788 * X) + (-3.87) 
```
